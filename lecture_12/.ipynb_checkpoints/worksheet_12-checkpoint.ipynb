c) In your own words, briefly explain:

- underfitting
- overfitting

and what signs to look out for for each.


Underfitting occurs when a model is too simple to capture the underlying patterns of the data. It might not have enough parameters or complexity to learn from the dataset, leading to poor performance both on training and new data. Signs of underfitting include high error rates on both the training set and test set, suggesting the model cannot generalize well even to the data it was trained on.

Overfitting, on the other hand, happens when a model is too complex, with too many parameters relative to the number of observations. It learns not only the underlying patterns but also the noise in the training data, making it perform exceptionally well on the training set but poorly on any new, unseen data. Signs of overfitting include a low error rate on the training data but a high error rate on the test data, indicating the model has learned the training data too well, including its anomalies and noise, and fails to generalize to new information.

e) Write a function to compute the Euclidean distance from it to all points in the dataset and pick the 3 closest points to it. In a scatter plot, draw a circle centered around the point with radius the distance of the farthest of the three points.

```python
def n_closest_to(example, n):
    dataset_points = np.array([(a, b) for a, b in zip(data['Attribute A'], data['Attribute B'])])
    distances = np.sqrt(np.sum((dataset_points - np.array(example))**2, axis=1))
    closest_indices = np.argsort(distances)[:n]
    return dataset_points[closest_indices], distances[closest_indices]

location = (0.5, 1)

closest_points, closest_distances = n_closest_to(location, 3)

radius = np.max(closest_distances)

_, axes = plt.subplots()
axes.scatter(data["Attribute A"], data["Attribute B"], color='lightgray', label='Original Points')
axes.scatter(closest_points[:, 0], closest_points[:, 1], color='green', label='3 Closest Points')
axes.scatter([location[0]], [location[1]], color='red', label='New Point')
cir = plt.Circle(location, radius, fill=False, edgecolor='blue', linestyle='--', alpha=0.8)
axes.add_patch(cir)
axes.set_aspect('equal')
plt.legend()
plt.show()
```

f) Write a function that takes the three points returned by your function in e) and returns the class that the majority of points have (break ties with a deterministic default class of your choosing). Print the class assigned to this new point by your function.

```python
def majority(points):
    # extract closes points
    closest_classes = np.array([data['Class'][i] for i, _ in enumerate(data['Attribute A']) if [data['Attribute A'][i], data['Attribute B'][i]] in points.tolist()])
    counts = np.bincount(closest_classes)
    return np.argmax(counts) if counts.size else 0  

majority_class = majority(closest_points)

print(f"The majority class among the 3 closest points is: {majority_class}")
```

g) Re-using the functions from e) and f), you should be able to assign a class to any new point. In this exercise we will implement Leave-one-out cross validiation in order to evaluate the performance of our model.

For each point in the dataset:

- consider that point as your test set and the rest of the data as your training set
- classify that point using the training set
- keep track of whether you were correct with the use of a counter

Once you've iterated through the entire dataset, divide the counter by the number of points in the dataset to report an overall testing accuracy.

```python
count = 0

for i in range(len(data["Attribute A"])):
    actual_class = data["Class"][i]
    
    training_data = {
        "Attribute A": np.delete(data["Attribute A"], i),
        "Attribute B": np.delete(data["Attribute B"], i),
        "Class": np.delete(data["Class"], i)
    }
    
    test_point = (data["Attribute A"][i], data["Attribute B"][i])
    
    closest_points, _ = n_closest_to(test_point, 3)
    
    prediction = majority(closest_points)
    
    if prediction == actual_class:
        count += 1

overall_accuracy = count / len(data["Attribute A"])
print(f"Overall accuracy = {overall_accuracy:.2f}")
```

b) For K ranging from 1 to 20:

1. train a KNN on the training data
2. record the training and testing accuracy

Plot a graph of the training and testing set accuracy as a function of the number of neighbors K (on the same plot). Which value of K is optimal? Briefly explain.

```python
from sklearn.neighbors import KNeighborsClassifier
X, y = fetch_openml(name="mnist_784", version=1, return_X_y=True, as_frame=False)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

training_accuracies = []
testing_accuracies = []

for K in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=K)
    knn.fit(X_train, y_train)
    
    training_accuracy = knn.score(X_train, y_train)
    testing_accuracy = knn.score(X_test, y_test)
    
    training_accuracies.append(training_accuracy)
    testing_accuracies.append(testing_accuracy)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 21), training_accuracies, label='Training Accuracy')
plt.plot(range(1, 21), testing_accuracies, label='Testing Accuracy')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Accuracy')
plt.xticks(range(1, 21))
plt.title('KNN Accuracy vs. Number of Neighbors')
plt.legend()
plt.grid(True)
plt.show()
```

c) Using the best model from b), pick an image at random and plot it next to its K nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

optimal_k = 3

X, y = fetch_openml(name="mnist_784", version=1, return_X_y=True, as_frame=False)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

knn = KNeighborsClassifier(n_neighbors=optimal_k)

knn.fit(X_train, y_train)

random_index = np.random.randint(len(X_test))
random_image = X_test[random_index]

distances, indices = knn.kneighbors([random_image])

fig, axes = plt.subplots(1, optimal_k + 1, figsize=(10, 3))

axes[0].imshow(random_image.reshape(28, 28), cmap='gray')
axes[0].set_title("Random Test Image")
axes[0].axis('off')

for i, (distance, idx) in enumerate(zip(distances[0], indices[0]), start=1):
    neighbor_image = X_train[idx].reshape(28, 28)
    axes[i].imshow(neighbor_image, cmap='gray')
    axes[i].set_title(f"Neighbor {i}\nDistance: {distance:.2f}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

d) Using a dimensionality reduction technique discussed in class, reduce the dimensionality of the dataset before applying a KNN model. Repeat b) and discuss similarities and differences to the previous model. Briefly discuss your choice of dimension and why you think the performance / accuracy of the model has changed.

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

dimensionality_reduction = TruncatedSVD(n_components=50)  
knn_model = KNeighborsClassifier(n_neighbors=3)

model = make_pipeline(dimensionality_reduction, knn_model)

X, y = fetch_openml(name="mnist_784", version=1, return_X_y=True, as_frame=False)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model.fit(X_train, y_train)

training_accuracy = model.score(X_train, y_train)
testing_accuracy = model.score(X_test, y_test)
print(f"Training accuracy: {training_accuracy}, Testing accuracy: {testing_accuracy}")
```

The training accuracy of approximately 98.90% and testing accuracy of 97.66% suggest that the KNN classifier, when applied after dimensionality reduction, maintains high predictive performance. The choice of dimension reduction technique and the number of dimensions to reduce to depend on the trade-off between computational efficiency and model accuracy. Reducing the dimensionality can help alleviate the curse of dimensionality, which can impede KNN performance due to increased distance computations in higher-dimensional spaces. By representing the dataset in a lower-dimensional space, the distances between points can become more meaningful, often improving KNN's ability to classify unseen data effectively. The improved performance suggests that the dimensionality reduction technique used preserved significant information while discarding noise and less informative features, leading to a compact representation that's amenable to KNN classification.

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train['FamilySize'] = train['Parch'] + train['SibSp']
test['FamilySize'] = test['Parch'] + test['SibSp']

train['Title'] = train['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test['Title'] = test['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

train.ffill(inplace=True)
test.ffill(inplace=True)

# Convert categorical variables to numerical
train = pd.get_dummies(train, columns=['Sex', 'Embarked', 'Title'])
test = pd.get_dummies(test, columns=['Sex', 'Embarked', 'Title'])

# Feature scaling
features = ['Pclass', 'Age', 'Fare', 'FamilySize', 'Sex_male', 'Sex_female', 'Embarked_S', 'Embarked_C', 'Embarked_Q', 'Title_Mr', 'Title_Miss', 'Title_Mrs', 'Title_Master']
X = train[features]
y = train['Survived']
X_test = test[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)

# Choosing K
knn = KNeighborsClassifier()
param_grid = {'n_neighbors': [3, 5, 7, 9, 11]}
grid_search = GridSearchCV(knn, param_grid, cv=5)
grid_search.fit(X_scaled, y)
best_k = grid_search.best_params_['n_neighbors']

# Training with best K
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_scaled, y)

predictions = knn.predict(X_test_scaled)

submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': predictions})
submission.to_csv('titanic_predictions.csv', index=False)
```
